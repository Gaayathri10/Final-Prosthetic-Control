```python
import os
import pandas as pd


def load_all_emg_data(raw_dir="data/raw"):

    #loads all CSV files from the raw data dir, adds a label column based on filename, and concatenates into a single DataFrame


    all_data = []

    for filename in os.listdir(raw_dir):
        if filename.endswith(".csv"):

            filepath = os.path.join(raw_dir, filename)

            # Load CSV
            df = pd.read_csv(filepath)

            # Extract label from filename
            # Example: "hand_open.csv" → "hand_open"
            label = filename.replace(".csv", "")

            df["label"] = label

            all_data.append(df)

            print(f"Loaded: {filename}")

    # Concatenate all gesture recordings
    combined_df = pd.concat(all_data, ignore_index=True)

    return combined_df
```
